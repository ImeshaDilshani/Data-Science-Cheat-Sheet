```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]  # Generate three datasets with different standard deviations
```

```python
# Plotting the box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['A', 'B', 'C'], notch=True, patch_artist=True, showmeans=True)
plt.title('Box Plot of Datasets A, B, C')
plt.xlabel('Dataset')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
