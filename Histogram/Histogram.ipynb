```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random data (normal distribution)
data = np.random.randn(1000)
```

```python
#ploting the hisogram
plt.figure(figsize=(8,6))
```

```python
plt.hist(data,bins=30, color='skyblue', alpha=0.7) # Plot histogram with 30 bins, skyblue color, and transparency
```

```python
plt.title('Histogram of Random Data')
```

```python
plt.xlabel('Values')
```

```python
plt.ylabel('Frequency')
```

```python
plt.grid(True)
```

```python
plt.show()
```
