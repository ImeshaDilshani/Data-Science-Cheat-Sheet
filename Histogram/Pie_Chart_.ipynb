```python
import matplotlib.pyplot as plt
```

```python
# Prepare categorical labels and corresponding values
labels = ['A', 'B', 'C', 'D']
sizes = [25, 35, 20, 20]  # Proportional sizes of each category (sums to 100 for percentage)
```

```python
# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'yellowgreen',
'orange', 'tomato'])
plt.title('Pie Chart of Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
