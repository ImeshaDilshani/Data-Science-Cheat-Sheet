```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Create a 2D NumPy array with random data
# Let's create 100 points with x and y coordinates
num_points = 100
```

```python
# Random x values between 0 and 100
x_values = np.random.randint(0, 100, num_points)
```

```python
# Random y values between 0 and 100
y_values = np.random.randint(0, 100, num_points)
```

```python
# Combine x and y values into a 2D array
data = np.column_stack((x_values, y_values))
```

```python
# Plotting the scatter plot
plt.figure(figsize=(8, 6))
```

```python
plt.scatter(data[:, 0], data[:, 1], color='blue', alpha=0.6)  # Scatter plot of x vs y
```

```python
plt.title('Scatter Plot of Random Data')
```

```python
plt.xlabel('X values')
plt.ylabel('Y values')
```

```python
plt.grid(True)
```

```python
plt.show()
```
