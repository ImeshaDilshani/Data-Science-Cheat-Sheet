```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate x values (e.g., time steps)
x = np.linspace(0, 10, 100)  # Create 100 evenly spaced values from 0 to 10
```

```python
# Generate y values for multiple lines (sine and cosine functions)
y1 = np.sin(x)  # Sine of x
y2 = np.cos(x)  # Cosine of x
```

```python
# Plotting multiple lines on the same graph
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='Sine', color='blue', linestyle='-', linewidth=2)  # Plot sine curve
plt.plot(x, y2, label='Cosine', color='red', linestyle='--', linewidth=2)  # Plot cosine curve
```

```python
plt.title('Multiple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()  # Show legend based on 'label' parameter
plt.grid(True)
plt.show()
```
